```python
import matplotlib.pylab as plt
import matplotlib as mpl
import numpy as np
import h5py
from scipy.special import erf
mpl.style.use('./plotting.mplstyle')
from makeCorner import *
```

```python
# Load AR results: Spins
chi_cost_data = h5py.File("./../data/ar_chi_cost_summary.hdf","r")
```

```python
plot_data = {
    'chi_std':{'data':chi_cost_data['posterior/ar_chi_std'][()],'plot_bounds':(0,4),'label':r'$\sigma_\chi$'},
    'chi_tau':{'data':np.log(chi_cost_data['posterior/ar_chi_tau'][()]),'plot_bounds':(-3,3),'label':r'$\ln\tau_\chi$'},
    'cost_std':{'data':chi_cost_data['posterior/ar_cost_std'][()],'plot_bounds':(0,4),'label':r'$\sigma_{\cos\theta}$'},
    'cost_tau':{'data':np.log(chi_cost_data['posterior/ar_cost_tau'][()]),'plot_bounds':(-5,18),'label':r'$\ln\tau_{\cos\theta}$'},
    'neff':{'data':np.log10(chi_cost_data['posterior/nEff_inj_per_event'][()]),'plot_bounds':(0,2.5),
            'label':r'$\log_{10}(N^\mathrm{inj}_\mathrm{eff}/N_\mathrm{obs})$'},
    'min_neff':{'data':chi_cost_data['posterior/min_log_neff'][()],'plot_bounds':(-0.5,2.7),
                'label':r'$\mathrm{min}\log_{10}(N_\mathrm{eff}^\mathrm{samp})$'},
}

fig = plt.figure(figsize=(10,10))
plot_corner(fig,plot_data,'#3182bd')
plt.subplots_adjust(hspace=0.1,wspace=0.15)
plt.savefig('figure_18.pdf',bbox_inches='tight',dpi=200)
plt.show()
```
